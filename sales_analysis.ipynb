# Sales Data Analysis
Week 3 Project: Introduction to Data Analysis  
This notebook loads sales data, performs basic cleaning, and calculates key metrics.

# 1. Load Data

```python
import pandas as pd

# Read the CSV file – adjust the path if needed
df = pd.read_csv(r'C:\Users\harsh\OneDrive\Desktop\Arena Developer Internship\Week_03_Sales_Analysisr_Project\sales_data.csv')

# Display first 5 rows
print("First 5 rows of the dataset:")
print(df.head())
print("\n" + "="*50 + "\n")
```

# 2. Explore Data

```python
# Check shape
print(f"Dataset shape: {df.shape[0]} rows, {df.shape[1]} columns\n")

# Get basic info
print("Data info:")
df.info()
print("\n" + "="*50 + "\n")
```

# 3. Clean Data

```python
# Check for missing values
missing = df.isnull().sum()
print("Missing values per column:")
print(missing[missing > 0] if any(missing > 0) else "No missing values found.")

# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"\nDuplicate rows: {duplicates}")
print("\n" + "="*50 + "\n")
```

# 4. Analyze Sales

```python
# Metric 1: Total Revenue
total_revenue = df['Total_Sales'].sum()
print(f"Total Revenue: Rs{total_revenue:,.2f}")

# Metric 2: Best-selling product by quantity
best_selling_qty = df.groupby('Product')['Quantity'].sum().idxmax()
max_qty = df.groupby('Product')['Quantity'].sum().max()
print(f"Best-selling product (by quantity): {best_selling_qty} (total {max_qty} units)")

# Metric 3: Best-selling product by revenue
best_selling_rev = df.groupby('Product')['Total_Sales'].sum().idxmax()
max_rev = df.groupby('Product')['Total_Sales'].sum().max()
print(f"Best-selling product (by revenue): {best_selling_rev} (total Rs{max_rev:,.2f})")

# Metric 4: Total sales per region
region_sales = df.groupby('Region')['Total_Sales'].sum().sort_values(ascending=False)
print("\nTotal Sales by Region:")
for region, sales in region_sales.items():
    print(f"  {region}: Rs{sales:,.2f}")
```

# 5. Create Basic Report

```python
print("\n" + "="*50)
print("SALES ANALYSIS REPORT SUMMARY")
print("="*50)
print(f"Date range: {df['Date'].min()} to {df['Date'].max()}")
print(f"Total transactions: {len(df)}")
print(f"Total Revenue: Rs{total_revenue:,.2f}")
print(f"Best-selling product (by quantity): {best_selling_qty} ({max_qty} units)")
print(f"Best-selling product (by revenue): {best_selling_rev} (Rs{max_rev:,.2f})")
print("\nRegion Performance:")
for region, sales in region_sales.items():
    print(f"  {region}: Rs{sales:,.2f}")
print("="*50)
```
